# INTERPOLAÃ‡Ã•ES

```python
def newton_interpolation(X, Y, x):
    n = len(X)
    f = [[0 for _ in range(n)] for _ in range(n)]
    for i in range(n):
        f[i][0] = Y[i]

    for j in range(1, n):
        for i in range(n - j):
            f[i][j] = (f[i + 1][j - 1] - f[i][j - 1]) / (X[i + j] - X[i])

    result = f[0][0]
    mult = 1.0
    for i in range(1, n):
        mult *= (x - X[i - 1])
        result += f[0][i] * mult

    return result
```

```python
colors = {
    'linear': 'yellow',
    'nearest': 'green',
    'zero': 'red',
    'quadratic': 'purple',
    'cubic': 'brown',
    'newton': 'pink'
}
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d
from scipy.interpolate import lagrange

def interpolate_all(df, df_test, n=5, figsize=(8, 10)):  # You can adjust figsize as needed
    X = df['X'].values.tolist()
    Y = df['Y'].values.tolist()

    X_test = df_test['X'].values
    Y_test = df_test['Y'].values

    new_X = np.arange(min(X), max(X), n)

    linear_interpolator = interp1d(X, Y, fill_value="extrapolate")
    interpolated_Y = linear_interpolator(new_X)

    methods = ['linear', 'nearest', 'zero', 'quadratic', 'cubic']

    interpolators = [interp1d(X, Y, kind=method) for method in methods]
    interpolated_values = [interpolator(new_X) for interpolator in interpolators]

    lagrange_poly = lagrange(X, Y)
    lagrange_Y = lagrange_poly(new_X)

    newton_Y = [newton_interpolation(X, Y, x_val) for x_val in new_X]

    # Set the figure size
    plt.figure(figsize=figsize)

    plt.plot(X, Y, '*', markersize=6, label='Original Data')
    plt.plot(X_test, Y_test, 'o', markersize=5, label='Test Data',  markerfacecolor='none', markeredgecolor='red')

    for i, method in enumerate(methods):
        plt.plot(new_X, interpolated_values[i], label=method)

    #plt.plot(new_X, lagrange_Y, '--', label='Lagrange')
    plt.plot(new_X, newton_Y, ':', label='Newton')


    plt.ylim([0, max(Y) + 10])

    plt.legend()

    #plt.xlim([0,  max(X)+5])

    plt.show()
```

```python
def r_squared(y_actual, y_predicted):
    y_mean = np.mean(y_actual)
    ss_total = np.sum((y_actual - y_mean) ** 2)
    ss_res = np.sum((y_actual - y_predicted) ** 2)
    return 1 - (ss_res / ss_total)

def interpolate_multiple(df, df_test, n=5, figsize=(8, 8)):
    X = df['X'].values.tolist()
    Y = df['Y'].values.tolist()

    X_test = df_test['X'].values
    Y_test = df_test['Y'].values

    new_X = np.arange(min(X), max(X), n)
    r2_scores = {}

    methods = ['linear', 'nearest', 'zero', 'quadratic', 'cubic', 'newton']

    for method in methods:
        plt.figure(figsize=figsize)

        if method == 'newton':
            interpolated_values = [newton_interpolation(X, Y, x_val) for x_val in new_X]
            y_pred_test = [newton_interpolation(X, Y, x_val) for x_val in X_test]
        else:
            interpolator = interp1d(X, Y, kind=method, fill_value="extrapolate")
            interpolated_values = interpolator(new_X)
            y_pred_test = interpolator(X_test)

        r2_scores[method] = r_squared(Y_test, y_pred_test)

        plt.plot(new_X, interpolated_values, color=colors[method], label=method)
        plt.title(f'Interpolation Method: {method} - R^2: {r2_scores[method]:.4f}')
        plt.ylim([0, max(Y) + 10])
        plt.plot(X, Y, '*', markersize=4, label='Data', color='blue')
        plt.plot(X_test, Y_test, 'o', markersize=5, label='Test Data',  markerfacecolor='none', markeredgecolor='red')
        plt.legend()
        plt.grid(True, alpha=0.3)
        plt.show()

    #return r2_scores

interpolate_multiple(train, test, .1)
```

```python
M = df['X']
C = df['Y']
```

```python
p2 = np.polyfit(M,C,2)
p3 = np.polyfit(M,C,3)
p4 = np.polyfit(M,C,4)
p5 = np.polyfit(M,C,5)
```

```python
C22 = p2[0]*M**2 + p2[1]*M    + p2[2] # modelo quadrÃ¡tico
C23 = p3[0]*M**3 + p3[1]*M**2 + p3[2]*M    + p3[3] # modelo cÃºbico
C24 = p4[0]*M**4 + p4[1]*M**3 + p4[2]*M**2 + p4[3]*M    + p4[4] # modelo de quarta ordem
C25 = p5[0]*M**5 + p5[1]*M**4 + p5[2]*M**3 + p5[3]*M**2 + p5[4]*M + p5[5] # modelo de quinta ordem

# plotagem
fig, ax = plt.subplots(constrained_layout=True)
ax.grid(True, alpha=0.4)

ax.plot(M,C,  'ko',ms=6, mfc='black', label='dado')
ax.plot(M,C22,'rs',ms=8, mfc='None', label='Ord2')
ax.plot(M,C23,'gd',ms=8, mfc='None', label='Ord3')
ax.plot(M,C24,'cv',ms=8, mfc='None', label='Ord4')
ax.plot(M,C25,'m*',ms=8, mfc='None', label='Ord5')

ax.set_xlabel('Weight')
ax.set_ylabel('Body Fat')
ax.legend(loc='best');
```
