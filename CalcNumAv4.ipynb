## Sun

$$ (x-10)^2 + (y-25)^2 = 25 $$


```python
import numpy as np
import matplotlib.pyplot as plt

# Define circle's parameters
center_x = 10
center_y = 25
radius = np.sqrt(25)

# Generate x values within the circle's domain
x = np.linspace(center_x - radius, center_x + radius, 1000)

# Compute y values from circle equation
y_positive = center_y + np.sqrt(radius**2 - (x-center_x)**2)
y_negative = center_y - np.sqrt(radius**2 - (x-center_x)**2)

# Plotting
plt.figure(figsize=(10,6))

# Plot both the positive and negative parts of the circle in orange color
plt.plot(x, y_positive, color='orange', label='Circle: (x-10)^2 + (y-25)^2 = 18')
plt.plot(x, y_negative, color='orange')

# Setting title, labels, and limits
plt.title('Second Equation')
plt.xlabel('x')
plt.ylabel('y')
plt.xlim([0, 60])
plt.ylim([0, 35])

plt.grid(True)
plt.legend()
plt.show()
```

## Bird

```python
def bird_plot(x_pos, y_pos, n, COLOR='brown'):
  x = np.linspace(0, 60, 1000)
  p_x = np.abs(x - x_pos) * n + y_pos

  return plt.plot(x, p_x, color=COLOR, label=f'p(x) = |x-{x_pos}| x 0.3 + {y_pos}')
```

```python
def bird(x_pos, y_pos):
  x = np.linspace(0, 60, 1000)

  # Plotting
  plt.figure(figsize=(10,6))

  bird_plot(x_pos, y_pos, 0.3)
  bird_plot(x_pos, y_pos-0.5, 0.5)

  # Setting title, labels, and limits
  plt.title('Bird')
  plt.xlabel('x')
  plt.ylabel('p(x)')
  plt.xlim([0, 60])
  plt.ylim([0, 35])

  plt.grid(True)
  plt.legend()
  plt.show()
```

## Regra do Trapézio


```python
# 1/3 Simpson

import numpy as np
import matplotlib.pyplot as plt
import sympy as sy
import scipy as sp
from scipy import interpolate
from scipy import integrate
%matplotlib inline

# Representação geométrica da regra do trapézio (simples)

def f(x):
    return -x**2 + 15*x - 10

x = np.linspace(0.5,10,100)
y = f(x)

plt.plot(x,y, label = 'f(x)')
plt.plot([2,8], [f(2),f(8)], 'r', label = 'f1(x)')
plt.plot(2,f(2), 'o', label = 'f(a)')
plt.plot(8,f(8), 'o', label = 'f(b)')
plt.plot(np.mean([2,8]), np.mean([f(2),f(8)]), 'o', label = 'Altura Média')
plt.fill([2, 2, 8, 8],[0, f(2), f(8), 0],color='r',alpha=0.4, label = 'Área do Trapézio')
plt.legend()
plt.axvline(x=0,color='k',linewidth=0.6,linestyle='--')
plt.axhline(y=0,color='k',linewidth=0.6,linestyle='--');
```

```python
# Representação geométrica da aplicação múltipla da regra do trapézio
# É possível observar, visualmente, que houve uma redução do erro do resultado em comparação com o caso anterior

xd = np.array([2,5,8])
yd = f(xd)

plt.plot(x,y, label = 'f(x)')
plt.plot(xd,yd, 'r')
plt.plot(2,f(2), 'o', label = 'f(x0)')
plt.plot(5,f(5), 'o', label = 'f(x1)')
plt.plot(8,f(8), 'o', label = 'f(x2)')
plt.fill([2, 2, 5, 5],[0, f(2), f(5), 0], color='r', alpha=0.6, label = 'Área do Trapézio 1')
plt.fill([5, 5, 8, 8],[0, f(5), f(8), 0], color='r', alpha=0.3, label = 'Área do Trapézio 2')
plt.legend()
plt.axvline(x=0,color='k',linewidth=0.6,linestyle='--')
plt.axhline(y=0,color='k',linewidth=0.6,linestyle='--');
```

## Simpson 1/3

f(x) = 0.4 \sin(x) + 5

$$
\int_{0}^{60} (0.4 \sin(x) + 5) \, dx
$$
