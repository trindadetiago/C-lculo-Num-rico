
# 1. Introdução do Contexto
Chernobyl foi palco de um dos piores acidentes nucleares da história em 1986. A liberação de material radioativo resultou em contaminação de larga escala, tornando partes da região inabitáveis. Vamos investigar o decaimento dos resíduos nucleares e o tempo necessário para que a área se torne segura novamente.


---

# 2. Apresentação da Fórmula Matemática
Para modelar o decaimento dos resíduos nucleares, podemos utilizar a lei do decaimento radioativo:

###dN/dt = -λN



*   N: quantidade de substância radioativa
*   λ: constante de decaimento (específica para cada isótopo)



---
# 3. Resolução do Problema Através dos Métodos Numéricos

Vamos aplicar O Método de Newton-Raphson para resolver essa equação diferencial. Podemos fazer isso transformando a equação diferencial em uma equação algébrica e, em seguida, aplicando o método.


---




## Etapa 1: Definir a Equação Diferencial
Vamos considerar o decaimento do césio-137, que foi um dos principais isótopos liberados no acidente de Chernobyl. A constante de decaimento para o césio-137 é de aproximadamente
0,6931 por ano.

A equação diferencial do decaimento é:

###dN/dt = −λN

N é a quantidade de substância radioativa e
λ é a constante de decaimento.

## Etapa 2: Discretização da Equação Diferencial
Para aplicar esse método numéricos precisamos transformar a equação diferencial em uma equação algébrica. Vamos usar o Método de Newton-Raphson para resolver essa equação.

Primeiro, vamos discretizar a equação diferencial usando o método de diferenças finitas:

###N[i+1] = N[i]  −λNt


onde N[i] é a quantidade de substância no tempo t[i] e t é o intervalo de tempo.


##Etapa 3: Aplicar o Método de Newton-Raphson
Vamos aplicar o Método de Newton-Raphson para resolver a equação discretizada e encontrar o valor de
N ao longo do tempo.

##Etapa 4: Análise dos Resultados
Finalmente, vamos analisar os resultados, plotar o gráfico do decaimento ao longo do tempo e discutir o que isso significa para a habitabilidade da área afetada.

Vamos começar com as etapas 2 e 3, discretizando a equação e aplicando o Método de Newton-Raphson.

```python
import numpy as np
import matplotlib.pyplot as plt
```

###Definindo os parametros

```python
#meia vida de 30 anos
lambda_cesium_137 = 0.0230433552
initial_quantity = 100
time_step = 1
total_time = 100
```

```python
def finite_diference_decay(quantity, lambda_value, delta_t):
    return quantity - lambda_value * quantity * delta_t
```

### Simulação do decaimento

```python
quantities = [initial_quantity]
for i in range(total_time):
    new_quantity = finite_diference_decay(quantities[-1], lambda_cesium_137, time_step)
    quantities.append(new_quantity)
```

```python
# Plotando o gráfico do decaimento
plt.figure(figsize=(10, 6))
plt.plot(quantities, label="Césio-137")
plt.title("Decaimento Radioativo do Césio-137 em Chernobyl")
plt.xlabel("Tempo (anos)")
plt.ylabel("Quantidade de Substância Radioativa")
plt.legend()
plt.grid(True)
plt.show()
```

Dada a constante de decaimento do césio-137, podemos ver que a substância ainda está presente após 100 anos, embora em quantidades significativamente reduzidas.
Para entender completamente a habitabilidade da área, seria necessário considerar outros isótopos e fatores, mas este modelo fornece uma visão simplificada do decaimento de um dos principais contaminantes.

```python

# Função para plotar a dispersão circular
def plot_circular(dispersion_data, isotope_name, time_slices, color_map):
    plt.figure(figsize=(8, 8))

    for time_slice in time_slices:
        data = dispersion_data[time_slice]
        angles = np.linspace(0, 2 * np.pi, len(data), endpoint=False)
       # plt.polar(angles, data, label=f"Ano {time_slice}", color=color_map[time_slice])

    center_data = np.mean(dispersion_data, axis=0)
    center_angles = np.linspace(0, 2 * np.pi, len(center_data), endpoint=False)

    diagonal_values = [max(dispersion_data[time_slice]) for time_slice in time_slices]
    center_diagonal = max(center_data)

    for i, diagonal in enumerate(diagonal_values):
        circle_radius = diagonal / 2
        angles = np.linspace(0, 2 * np.pi, 360)
        radii = [circle_radius] * 360
        plt.polar(angles, radii, color=color_map[time_slices[i]], linestyle='solid', alpha=1.0,
                  label=f"Ano {time_slices[i]} - Diagonal")

    center_circle_radius = center_diagonal / 2
    angles = np.linspace(0, 2 * np.pi, 360)
    radii = [center_circle_radius] * 360
    plt.polar(angles, radii, color='gray', linestyle='solid', alpha=1.0,
              label="Diagonal da Dispersão Centralizada")

    plt.title(f"Dispersão Circular do {isotope_name} em Chernobyl em Diferentes Intervalos de Tempo")
    plt.legend()
    plt.grid(True)
    plt.show()

# Plotagem das dispersões para Césio-137 e Estrôncio-90
plot_circular(dispersion_cesium, "Césio-137", [0, 10, 50, 100], color_map)
plot_circular(dispersion_strontium, "Estrôncio-90", [0, 10, 50, 100], color_map)
```

```python

import numpy as np
import matplotlib.pyplot as plt



# Função que simula a dispersão e calcula o valor da função
def simulate_dispersion(quantities, dispersion_std_dev_initial, dispersion_spread_rate, total_time, guess):
    dispersion_data = []
    for i in range(total_time + 1):
        std_dev = dispersion_std_dev_initial + i * dispersion_spread_rate
        dispersion = quantities[i] * np.exp(-(np.arange(-10, 11)**2) / (2 * std_dev**2))
        dispersion_data.append(dispersion)

    # Calcula o valor da função para o chute inicial dado
    value_at_guess = dispersion_data[guess]
    return value_at_guess

# Chute inicial para o método de Newton-Raphson
initial_guess = 50

# Implementação do método de Newton-Raphson para encontrar um valor que resulte em y = 0.1
def newton_raphson_method(func, initial_guess, target_value, tolerance=1e-6, max_iterations=100):
    x = initial_guess
    for _ in range(max_iterations):
        f_x = func(x)
        if abs(f_x - target_value) < tolerance:
            return x
        f_prime_x = 0  # A derivada da função constante é zero
        x = x - (f_x - target_value) / f_prime_x
    return None  # Retorna None se o valor desejado não for encontrado

# Aplicando o método de Newton-Raphson à função simulate_dispersion para encontrar um valor que resulte em y = 0.1
target_value = 0.1
result = newton_raphson_method(lambda x: simulate_dispersion(quantities_cesium, dispersion_std_dev_initial, dispersion_spread_rate, total_time, x), initial_guess, target_value)

if result is not None:
    print(f"Valor encontrado: {result}")
else:
    print(f"Não foi possível encontrar um valor que resulte em {target_value}.")

# Plotando o gráfico da dispersão
dispersion_at_result = simulate_dispersion(quantities_cesium, dispersion_std_dev_initial, dispersion_spread_rate, total_time, int(result))
plt.plot(np.arange(-10, 11), dispersion_at_result)
plt.axhline(y=target_value, color='r', linestyle='--', label='y = 0.1 (alvo)')
plt.title("Dispersão com Valor Alvo y = 0.1")
plt.xlabel("Distância da Origem (unidades de distância)")
plt.ylabel("Valor de Dispersão")
plt.legend()
plt.grid(True)
plt.show()
```
